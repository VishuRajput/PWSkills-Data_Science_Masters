```python
#OVERFLOWERROR - when the value is too high or long to print it will give over flow error.
import logging
logging.basicConfig(filename = 'Arrith.log', level=logging.DEBUG)
import math
try:
    logging.info("The exponential value is")
    logging.info('{}'.format(math.exp(1000)))
except OverflowError as e:
    logging.error(e)
logging.shutdown()
```

```python
#Problem 4

#LookupError - these are basically used to look for any wrong key being used to access dictionary or trying to access invalid index.

#Keyerror
import logging
logging.basicConfig(filename = 'lookup.log', level=logging.DEBUG)
try:
    d={'key1':'vishu','key2':'chib'}
    print(d['key3'])
except KeyError as e:
    logging.error('Trying to access a wrong key {}'.format(e))
    
#Indexerror

try :
    l=['vishu','chib','data','Science']
    print(l[4])
except IndexError as e:
    logging.error('Trying for invalid index {}'.format(e))
```
