```python
#Problem 2

from flask import Flask

app = Flask(__name__)

@app.route("/")
def hello_world():
    return "<h1>Hello, World! Welcome to my Local Server</h1>"



if __name__=="__main__":
    app.run(host="0.0.0.0")
```

#Problem 3

App routing is used to map the specific URL with the associated function that is intended to perform some task. 

It is used to access some particular page like Flask Tutorial in the web application.

```python
#Problem 4

from flask import Flask

app = Flask(__name__)

@app.route("/Welcome")
def Welcome():
    return "<h1>Welcome to ABC Corporation</h1>"

@app.route("/")
def Details():
    return '<h1>Company Name : ABC Corporation<br></br>Location : India<br></br>Contact Detail : 999-999-9999</h1>'

if __name__=="__main__":
    app.run(host="0.0.0.0")
```

```python
#Problem 5

#To build a URL to a specific function, use the url_for() function. It accepts the name of the function as its first argument and any number of keyword arguments, each corresponding to a variable part of the URL rule.


from flask import Flask,redirect, url_for,request

app = Flask(__name__)

@app.route("/accountType/<Type>")
def accountType(Type):
    return '<h1>This is a %s account</h1>' % Type

@app.route("/user/<name>")
def userType(name):
    return redirect(url_for('accountType',Type = name))


if __name__=="__main__":
    app.run(host="0.0.0.0")
    

```
