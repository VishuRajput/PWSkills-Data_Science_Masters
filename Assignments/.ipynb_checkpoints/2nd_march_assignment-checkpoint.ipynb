#Problem 1

Matplotlib is a cross-platform, data visualization and graphical plotting library (histograms, scatter plots, bar charts, etc) for Python and its numerical extension NumPy. As such, it offers a viable open source alternative to MATLAB

Plots :
- plot(x, y)
- scatter(x, y)
- bar(x, height)
- stem(x, y)
- step(x, y)

#Problem 2

A scatter plot is a type of plot in which individual data points are represented as markers on a two-dimensional Cartesian coordinate system. It is used to visualize the relationship or correlation between two numerical variables.

```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))

plt.scatter(x,y)
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.title('Scatter Plot')
plt.show()
```

#Problem 3

The subplot() function in Matplotlib is used to create a grid of subplots within a single figure. It allows you to divide the figure into multiple smaller axes (subplots) and arrange them in rows and columns

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.array([0, 1, 2, 3, 4, 5])

y1 = np.array([0, 100, 200, 300, 400, 500])
y2 = np.array([50, 20, 40, 20, 60, 70])
y3 = np.array([10, 20, 30, 40, 50, 60])
y4 = np.array([200, 350, 250, 550, 450, 150])

plt.figure(figsize=(10, 10))

plt.subplot(2,2,1)
plt.plot(x,y1,'r')
plt.title('Line 1')

plt.subplot(2,2,2)
plt.plot(x,y2,'b')
plt.title('Line 2')

plt.subplot(2,2,3)
plt.plot(x,y3,'y')
plt.title('Line 3')

plt.subplot(2,2,4)
plt.plot(x,y4,'g')
plt.title('Line 4')

plt.show()
```

#Problem 4

A bar chart is used when you want to show a distribution of data points or perform a comparison of metric values across different subgroups of your data. From a bar chart, we can see which groups are highest or most common, and how other groups compare against the others.

```python
import numpy as np
import matplotlib.pyplot

company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])


plt.bar(company,profit)
plt.xlabel('COMPANY NAME')
plt.ylabel('PROFIT')
plt.title('Company Profit Plot')
plt.show()

plt.barh(company,profit)
plt.xlabel('COMPANY NAME')
plt.ylabel('PROFIT')
plt.title('Company Profit Plot')
plt.show()
```

#Problem 5

Box plots are used to show distributions of numeric data values, especially when you want to compare them between multiple groups. They are built to provide high-level information at a glance, offering general information about a group of data's symmetry, skew, variance, and outliers.

```python
import numpy as np
import matplotlib.pyplot as plt

box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)

data = (box1,box2)
plt.boxplot(data)
plt.show()
```
