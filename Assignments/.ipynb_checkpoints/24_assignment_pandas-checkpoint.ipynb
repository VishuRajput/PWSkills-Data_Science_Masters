```python
#Problem 4

data = {'Text':['Python','This is pwskills class','This is Data Science Course']}

df3 = pd.DataFrame(data)
ls=[]

for i in df3['Text']:
    ls.append(len(i.split()))
    
df3['Word_count'] = ls
df3
```

```python
#Problem 7

data = {'Email': ['vishu@gmail.com','chib@yahoo.com','sudh@gmail.com']}
df4 = pd.DataFrame(data)

ls = []
for i in df4['Email']:
    ls.append(i.split('@'))
    
df4['Username'] = [i[0] for i in ls]

df4
```
